```python

#This will not run on online IDE
import requests
from bs4 import BeautifulSoup
  
URL = "https://coinranking.com/"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'lxml') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
l= []
for i in soup.findAll('a', attrs = {'class':'profile__link'}):
    l.append(i.text)
print(l)
```

```python

for i in soup.findAll('td', attrs = {'class':"valuta valuta--light"}):
    print(i)
```
